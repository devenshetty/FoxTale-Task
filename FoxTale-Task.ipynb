## Developing the offer logic

```python
final_table = order_export_test.copy()
final_table.head()
```

```python
final_table['price3'].replace(0,float('inf'), inplace=True)
```

```python
def minimum_prices(row):
    quantity_columns = ['Quantity of SKU 1', 'Quantity of SKU 2', 'Quantity of SKU 3']
    price_columns = ['price1', 'price2', 'price3']
    prices_numeric = pd.to_numeric(row[['price1', 'price2', 'price3']], errors='coerce')
    lowest_priced_units = prices_numeric.nsmallest(2)
    row['L1'] = lowest_priced_units[0]
    row['L2'] = lowest_priced_units[1]
    return row
final_table = final_table.apply(minimum_prices, axis=1)
```

```python
final_table.head()
```

```python
final_table['price3'].replace(float('inf'), 0, inplace=True)
final_table['Total_Price'] = final_table['Quantity of SKU 1']*final_table['price1'] + final_table['Quantity of SKU 2']*final_table['price2'] + final_table['Quantity of SKU 3']*final_table['price3']
final_table['Total units sold'] = final_table[['Quantity of SKU 1', 'Quantity of SKU 2', 'Quantity of SKU 3']].apply(lambda row: row.sum(), axis=1)
```

```python
final_table
```

```python
def calculate_promotion_metrics(row):
    quantity_columns = ['Quantity of SKU 1', 'Quantity of SKU 2', 'Quantity of SKU 3']
    price_columns = ['price1', 'price2', 'price3']
    if row['Discount code'] == 'B2G2':
        if row['Total units sold'] >= 4:
            row['GMV_Free'] = row['L1']+row['L2']
            row['GMV_NonFree'] = row['Total_Price'] - row['GMV_Free']
            row['Free units'] = 2
            row['Non Free units'] = row['Total units sold'] - 2
    elif row['Discount code'] == 'BOGO':
        if row['Total units sold'] >= 2:
            row['GMV_Free'] = row['L1']
            row['GMV_NonFree'] = row['Total_Price'] - row['GMV_Free']
            row['Free units'] = 1
            row['Non Free units'] = row['Total units sold'] - 1
    else:
        row['GMV_Free'] = 0
        row['GMV_NonFree'] = row['Total_Price'] 
        row['Free units'] = 0
        row['Non Free units'] = row['Total units sold'] 
            
    return row

final_table = final_table.apply(calculate_promotion_metrics, axis=1)
```
